Importing our libs

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Getting and selecting our data

```python
data = pd.read_csv('./ds_salaries.csv').to_numpy()
currency = data[:, 6] == 'USD'
salaries = data[:, 5][currency]
experince_levels = data[:, 2][currency]
```

scattering salaries acording to experince level

```python
plt.scatter(x=experince_levels, y=salaries)
plt.xlabel('Experince Level')
plt.ylabel('Salaries')
```

Selecting the average of each experince level

```python
avg_SE_salaries = np.mean(salaries[experince_levels == 'SE'])
avg_MI_salaries = np.mean(salaries[experince_levels == 'MI'])
avg_EN_salaries = np.mean(salaries[experince_levels == 'EN'])
avg_EX_salaries = np.mean(salaries[experince_levels == 'EX'])
x = ["SE", "MI", "EN", "EX"]
y = [avg_SE_salaries, avg_MI_salaries, avg_EN_salaries, avg_EX_salaries]
plt.scatter(x,y)
```
